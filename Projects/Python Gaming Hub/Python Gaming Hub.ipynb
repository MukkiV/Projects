```python
# Python Game Hub

import random 

def play_riddle_game():
    riddles = {
        "What has keys but can't open locks?": "keyboard",
        "What can travel around the world while staying in the same corner?": "stamp",
        "What gets wetter as it dries?": "towel",
        "What has hands but canâ€™t clap?": "clock",
        "Iâ€™m tall when Iâ€™m young, and Iâ€™m short when Iâ€™m old. What am I?": "candle",
        "What has a neck but no head?": "bottle",
        "What has many teeth but cannot bite?": "comb"
    }
    
    score = 0
    total = 0
    
    riddle_list = list(riddles.items())
    random.shuffle(riddle_list)
    
    print("----Welcome to riddle game----")
    print("---Enter q to quit any time---\n")
    
    for que, ans in riddle_list:
        print(f"Question: {que}")
        user_input = input("Enter the answer (q to quit): ")
        while user_input.isdigit():
            print("Invalid entry")
            user_input = input("Enter the answer (q to quit): ")
        if user_input.lower() == "q":
            break
        elif user_input.lower() == ans:
            print("The answer is correct!")
            score += 1
            total += 1
        else:
            print("The answer is incorrect")
            print(f"The correct answer is {ans}")
            total += 1
        choice = input("Do you want to continue playing? (Yes/No): ")
        if choice.lower() != "yes":
            break
    
    print("----- Game Over -----")
    if total >0:
        percentage = (score/total)*100
        print(f"Your score: {score} out of {total} as {percentage:.2f}%")
    else:
        print("You didn't attempt any questions.")
    print("Thanks for playing!")

def play_arithmetic_puzzle():
    print("----Welcome to Maths game----")
    score = 0
    total = 0
    while True:
        num1 = random.randint(1,20)
        num2 = random.randint(1,20)
        operator = random.choice(["*", "+", "-", "/"]) 
        question = f"{num1} {operator} {num2}"
        answer = eval(question)
    
        print(f"Question: {question}")
        try:
            user_input = float(input("Enter the answer: "))
        except ValueError:
            print("You can enter only numbers")
            user_input = float(input("Enter the answer: "))
        if round(user_input, 2) == round(answer, 2):
            print("Your answer is correct!")
            score += 1
            total += 1
        elif user_input != answer:
            print("Your answer is incorrect")
            print(f"The correct answer is {answer: .2f}")
            total += 1 
        choice = input("Do you wish to play again (Yes/No): ")
        if choice.strip().lower() == "yes":
            continue
        elif choice.strip().lower() == "no":
            break
        else: 
            print("Invalid entry, exiting game.")
            break
    if total == 0:
        print("No games were played.")
    else:
        percentage = (score / total) * 100
        print(f"Your final score is: {score} out of {total} as {percentage:.2f}%")
    print("Thanks for playing!")


def play_gk_quiz():
    score = 0
    total = 0
    print("Welcome to General knowledge Quiz Game")
    
    def ask_category():
        science = {"Which gas do plants absorb during photosynthesis?": "Carbon dioxide", 
                    "What is the full name of the WWW?": "World Wide Web",
                    "How many bones are there in the adult human body?": "206"}
        maths = {"What is the only even prime number?": "2",
                 "What is the sum of the first 5 natural numbers?": "15",
                 "What is the name of the mathematical symbol Ï€?": "Pi"}
        
        geography =  {"What is the tallest mountain on Earth?": "Mount Everest", 
                      "What is the capital of France?": "Paris", 
                      "Which planet is known as the 'Red Planet'?": "Mars"}
        
        others = {"Which country won the most recent FIFA World Cup (as of 2022)?": "Argentina", 
                  "Who invented the light bulb?": "Thomas Edison", 
                  "In what year did India declare itself a Democratic and Republic state with the adoption of the Constitution?": "1950"}
        
        print("\nSelect a category:")
        print("1. Science")
        print("2. Maths")
        print("3. Geography")
        print("4. Others")
        print("5. Exit")
        choice = input("Enter your choice (1-5): ")
    
        isRunning = True
        
        while isRunning:
            if choice == "1":
                return science
            elif choice == "2":
                return maths
            elif choice == "3":
                return geography
            elif choice == "4":
                return others
            elif choice == "5":
                isRunning = False
            else:
                print("Invalid category, Try again!")
                choice = input("Enter your choice (1-5): ")
    
    def ask_question():
        exec_question(ask_category())
        display_score(total, score)
        
    def check_answer(user_answer, correct_answer):
        nonlocal score, total 
    
        if user_answer.strip().lower() == correct_answer.strip().lower():
            print("Your answer is correct")
            score += 1
            total += 1
        else:
            print("Your answer is incorrect")
            print(f"The correct answer is {correct_answer}")
            total += 1
        
    def display_score(total, score):
        print("\n----- Quiz Completed -----")
        print(f"Total Questions: {total}")
        print(f"Correct Answers: {score}")
        if total > 0:
            percentage = (score / total) * 100
            print(f"ðŸŽ¯ Your Score: {percentage:.2f}%")
        else:
            print("No questions were answered.")
        
    def exec_question(questions):
        for question, correct_answer in questions.items():
            print(f"Question: {question}")
            user_answer = input("Enter your answer (q to go back): ")
            if user_answer.strip().lower() == "q":
                break
            else:
                check_answer(user_answer, correct_answer)
    def play_game():
        while True:
            ask_question()
            while True:
                choice = input("Do you wish to play again (Yes/No): ")
                if choice.lower() == "yes":
                    ask_question()
                elif choice.lower() == "no":
                    return  
                else:
                    print("Invalid Entry. Please enter Yes or No.")
                    
    play_game()
    print("Thank you for playing! Goodbye ðŸ‘‹")

def play_rps_with_computer():  
    print("Welcome to Rock Paper Scissor Game")
    options = ["Rock", "Paper", "Scissor"]
    score = 0
    total = 0
    
    
    while True:
        user_input = input("Enter your choice (Rock/Paper/Scissor) (q to quit): ")
        if user_input.lower() == "q":
            break
        if user_input.title() not in options:
            print("Invalid entry! Try again")
            continue
        computer = random.choice(options)
        print(f"The computer choice is {computer}")
        print(f"Your choice is {user_input.title()}")
        if user_input.title() == computer:
            print("Its a tie")
            total += 1
        elif user_input.title() == "Paper" and computer == "Rock":
            print("Paper covers rock, You Won!")
            score += 1
            total += 1
        elif user_input.title() == "Rock" and computer == "Scissor":
            print("Rock crushes scissor, You Won!")
            score += 1
            total += 1
        elif user_input.title() == "Scissor" and computer == "Paper":
            print("Scissor cuts Paper, You Won!")
            score += 1
            total += 1
        elif user_input.title() == "Rock" and computer == "Paper":
            print("Paper covers rock, Computer Won!")
            total += 1
        elif user_input.title() == "Scissor" and computer == "Rock":
            print("Rock crushes scissor, Computer Won!")
            total += 1
        elif user_input.title() == "Paper" and computer == "Scissor":
            print("Scissor cuts Paper, Computer Won!")
            total += 1
    
    print("---Game Over---")
    if total == 0:
        print("No games were played.")
    else:
        percentage = (score / total) * 100
        print(f"Your final score is: {score} out of {total} as {percentage:.2f}%")

def play_rps_with_user():   
    options = ["Rock", "Paper", "Scissor"]
    user1_score = 0
    user2_score = 0
    total = 0
    
    print("---Welcome to Rock Paper Scissor game---")
    user1 = input("Hello user - 1, Enter your name: ")
    user2 = input("Hello user - 2, Enter your name: ")
    
    while True:
        user1_input = input(f"{user1} Enter your choice (Rock/Paper/Scissor) (q to quit): ")
        user2_input = input(f"{user2} Enter your choice (Rock/Paper/Scissor) (q to quit): ")
        if user1_input.lower() == "q" or user2_input.lower() == "q":
            break
        if user1_input.title() not in options and user2_input.title() not in options:
            print("Invalid entry by one or both users Try again")
            continue
        if user1_input.title() == user2_input.title():
            print("Its a tie")
            total += 1
        elif user1_input.title() == "Paper" and user2_input.title() == "Rock":
            print(f"Paper covers rock, {user1} Won!")
            user1_score += 1
            total += 1
        elif user1_input.title() == "Rock" and user2_input.title() == "Scissor":
            print(f"Rock crushes scissor, {user1} Won!")
            user1_score += 1
            total += 1
        elif user1_input.title() == "Scissor" and user2_input.title() == "Paper":
            print(f"Scissor cuts Paper, {user1} Won!")
            user1_score += 1
            total += 1
        elif user1_input.title() == "Rock" and user2_input.title() == "Paper":
            print(f"Paper covers rock, {user2} Won!")
            user2_score += 1
            total += 1
        elif user1_input.title() == "Scissor" and user2_input.title() == "Rock":
            print(f"Rock crushes scissor, {user2} Won!")
            user2_score += 1
            total += 1
        elif user1_input.title() == "Paper" and user2_input.title() == "Scissor":
            print(f"Scissor cuts Paper, {user2} Won!")
            user2_score += 1
            total += 1
    print("\n--- Game Over ---")
    print(f"Total Rounds: {total}")
    print(f"{user1}'s Score: {user1_score}")
    print(f"{user2}'s Score: {user2_score}")
    
    if user1_score > user2_score:
        print(f"{user1} is the overall winner!")
    elif user2_score > user1_score:
        print(f"{user2} is the overall winner!")
    else:
        print("It's a draw!")
    
    print("Thanks for playing!")

def play_word_guessing():
    word_list = ["python", "developer", "quiz", "science", "puzzle", "keyboard", "monitor"]
    secret_word = random.choice(word_list)
    print("Welcome to word guessing game")
    attempt = 1
    max_attempt = 5

    print(f"The clue for the secret word is {len(secret_word)} letter word")
    while attempt <=5:
        guess = input(f"Enter your guess attempt {attempt}/{max_attempt}: ").strip().lower()
        attempt += 1
        if guess == secret_word:
            print("Correct! You've guessed the word.")
            break
        else:
            print("Incorrect guess.")
        if attempt == 3:
            print(f"The secret word starts with : {secret_word[0]}")
    if guess != secret_word:
        print(f"You've used all attempts. The word was: '{secret_word}'.")

    print("Thanks for playing!")   

class GameMenu:
    @staticmethod
    def menu():
        print("\n=== Game Menu ===")
        print("1. Play Riddle Puzzle")
        print("2. Play Arithmetic Puzzle")
        print("3. Play General Knowledge Quiz")
        print("4. Play Rock-Paper-Scissor (Computer)")
        print("5. Play Rock-Paper-Scissor (User)")
        print("6. Play Word-Guessing Game")
        print("7. Exit")
        print("Enter a number between 1 and 7.")
        

def main():
    while True:
        try:
            GameMenu.menu()
            choice = input("Enter your choice: ")
            if choice == "1":
                play_riddle_game()
            elif choice == "2":
                play_arithmetic_puzzle()
            elif choice == "3":
                play_gk_quiz()
            elif choice == "4":
                play_rps_with_computer()
            elif choice == "5":
                play_rps_with_user()
            elif choice == "6":
                play_word_guessing()
            elif choice == "7":
                print("Exiting the game. Goodbye!")
                break
            else:
                print("Invalid choice. Please enter a number between 1 and 7.")
        except Exception as e:
            print(f"An error occurred: {e}")
        
        again = input("Do you want to play another game? (Y/N): ")
        if again.lower() != "y":
            print("Thank you for playing!")
            break

main()
```
